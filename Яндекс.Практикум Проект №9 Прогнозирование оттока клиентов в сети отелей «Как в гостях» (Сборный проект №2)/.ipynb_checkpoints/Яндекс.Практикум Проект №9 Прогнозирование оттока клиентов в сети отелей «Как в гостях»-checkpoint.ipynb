#### Предобработка данных

```python
# Ещё раз проверим данные на наличие пропусков
display(all(train_data.isna().sum()))
display(all(test_data.isna().sum()))
```

```python
# Изменим тип данных в столбцах adults, children и babies на int64
astype_columns = ['adults', 'children', 'babies']
train_data[astype_columns] = train_data[astype_columns].astype(dict.fromkeys(astype_columns, 'int64'))
test_data[astype_columns] = test_data[astype_columns].astype(dict.fromkeys(astype_columns, 'int64'))

display(train_data[astype_columns].dtypes)
display(test_data[astype_columns].dtypes)
```

```python
# Дополнительно проверим столбец id
display(train_data.id.duplicated().sum())
display(test_data.id.duplicated().sum())
```

Удалим неинформативные признаки, которые не несут ценности для прогноза:
- Признак `id` никак не связан с целевым признаком `is_canceled`
- Признак `total_nights` является суммой признаков `stays_in_weekend_nights` и `stays_in_week_nights`
- Признак `arrival_date_week_number` всегда будет линейно зависим с номером месяца, поэтому один из этих признаков нам точно не пригодится
- Признак `is_repeated_guest` является избыточным

```python
train_data.drop(['id', 'total_nights', 'arrival_date_week_number', 'is_repeated_guest'], axis=1, inplace=True)
test_data.drop(['id', 'total_nights', 'arrival_date_week_number', 'is_r epeated_guest'], axis=1, inplace=True)

display(train_data.columns)
display(test_data.columns)
```

Для кодирования такого признака как `country` лучше всего подойдёт способ кодирования `Ordinal Encoding`. Как известно `Ordinal Encoding` не подходит для логистической регрессии. Откажемся от использования этой модели в дальнейшем.

```python
encode_columns = ['arrival_date_month', 'meal', 'country', 'distribution_channel', 'reserved_room_type', 'customer_type']
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
encoder.fit(train_data[encode_columns])

train_data[encode_columns] = pd.DataFrame(encoder.transform(train_data[encode_columns]), columns=encode_columns)
test_data[encode_columns] = pd.DataFrame(encoder.transform(test_data[encode_columns]), columns=encode_columns)
```

```python
# Удалим строки в данных, предназначенных для тестирования модели, которые не удалось закодировать
display(test_data[test_data.country.isna()].shape)
test_data = test_data[~test_data.country.isna()]
display(test_data[test_data.country.isna()].shape)
```

**Промежуточный вывод:**
- Мы устранили несоответствия типов в данных
- Мы удалили неинформативные признаки, которые не несут ценности для прогноза
- Мы закодировали признаки при помощи способа кодирования `Ordinal Encoding`

```python
# Обработаем бронирования с детьми до 14 лет без взрослых
train_data = train_data[(train_data.adults != 0) | (train_data.children == 0)]
test_data = test_data[(test_data.adults != 0) | (test_data.children == 0)]
```

### Шаг 3. Формулировка ML-задачи на основе бизнес-задачи

```python
coeff_by_season = [1, 1.2, 1.4, 1.2]
price_by_type = {
    0: {'per_night': 1000, 'service': 400},
    1: {'per_night': 800, 'service': 350},
    2: {'per_night': 600, 'service': 350},
    3: {'per_night': 550, 'service': 150},
    4: {'per_night': 500, 'service': 150},
    5: {'per_night': 450, 'service': 150},
    6: {'per_night': 350, 'service': 150}
}


# Функция booking_revenue считает прибыль для одной брони с учётом отмен и различных сезонных коэффициентов 
def booking_revenue(row):
    price = price_by_type[int(row.reserved_room_type)]
    total_nights = int(row.stays_in_weekend_nights + row.stays_in_week_nights)
    start_date = datetime.datetime(int(row.arrival_date_year), 
                                   int(list(calendar.month_name).index(encoder.inverse_transform(row[encode_columns].values.reshape(1, -1))[0][0])),
                                   int(row.arrival_date_day_of_month))
    
    # Генератор для итерации по всем забронированным дням
    all_dates = (start_date + datetime.timedelta(days=x) for x in range(0, total_nights))
    
    if row['is_canceled']:
        revenue = -(price['per_night'] * coeff_by_season[(start_date.month % 12) // 3] + price['service'])
    else:
        # Учтём затраты на обслуживание при подготовке номера
        revenue = -price['service']
        second_night = False
        
        for date in all_dates:
            revenue += price['per_night'] * coeff_by_season[(date.month % 12) // 3] - price['service'] * second_night
            second_night = not second_night
            
    return round(revenue, 2)
```

```python
# Подсчитаем прибыль для каждого бронирования без внедрения депозитов
test_data['revenue'] = test_data.apply(booking_revenue, axis=1)
print(f'Прибыль отеля без внедрения депозитов — {test_data["revenue"].sum()}')
```

### Шаг 4. Разработка модели ML

Ранее мы изучили следующие модели: логистическая регрессия, дерево решений, случайный лес. Так как мы использовали способ кодирования `Ordinal Encoding`, который не совместим с логистической регрессией, будем использовать только дерево решений и случайный лес.

Рассмотрим изученные нами ранее метрики для задачи классификации — `accuracy`, `precision`, `recall`, `F1-score`, `ROC-AUC`:
- Известно, что `accuracy` зачастую не является оптимальным выбором в качестве метрики для проблем с дисбалансом классов
- Для нашей задачи важно минимизировать как `False Positive` так и `False Negative` предсказания модели. В самом деле, если модель ложно спрогнозировала отказ от брони и обязала клиента внести депозит (хотя на самом деле он ни в коем случае не собирался делать отмену) то он может и вовсе отказаться от предоставленных услуг и по итогу сеть потеряет очередного клиента. И наоборот, если модель не смогла спрогнозировать отказ от бронирования, то сеть в таком случае может потерпеть убытки. Поэтому мы не будем делать предпочтения на счёт выбора между `precision` и `recall` и остановимся на двух метриках — `F1-score` и `ROC-AUC`
- Метрику `F1-score` легче интерпретировать и объяснить её использование для бизнеса, чем метрику `ROC-AUC` поэтому будем использовать её в качестве основной метрики при выборе оптимальной модели. Тем не менее мы дополнительно посмотрим на кривую `ROC` и подсчитаем площадь по ней — `ROC-AUC` для наилучшей модели

Для борьбы с дисбалансом классов будем использовать параметр `class_weight`

```python
features = train_data.drop('is_canceled', axis=1)
target = train_data['is_canceled']

tree_param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(1, 51)),
    'min_samples_split': list(range(2, 6))
}

forest_param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(1, 51)),
    'n_estimators': list(range(1, 11)),
}

tree_search = GridSearchCV(DecisionTreeClassifier(random_state=state, class_weight='balanced'),
                           tree_param_grid,
                           scoring='f1',
                           n_jobs=-1,
                           refit=True,
                           cv=5,
                           verbose=3).fit(features, target)
forest_search = GridSearchCV(RandomForestClassifier(random_state=state, class_weight='balanced'),
                             forest_param_grid, 
                             scoring='f1',
                             n_jobs=-1,
                             refit=True,
                             cv=5,
                             verbose=3).fit(features, target)
```

```python
print(f'Гиперпараметры наилучшей модели дерева решений — {tree_search.best_estimator_.get_params()}')
print(f'Значение F1-меры для этой модели — {round(tree_search.best_score_, 4)}')
print()
print(f'Гиперпараметры наилучшей модели случайного леса — {forest_search.best_estimator_.get_params()}')
print(f'Значение F1-меры для этой модели — {round(forest_search.best_score_, 4)}')
```

Лучше всего себя показал модель `RandomForestClassifier` с гиперпараметрами `criterion` — `gini`, `max_depth` — `6` и `n_estimators` — `3`. Значение `F1-меры` этой модели при кросс-валидации `0.6654`. Выберем это модель для итогового тестирования

```python
# Функция add_deposit добавляет в прибыль депозит, если модель спрогнозировала отказ и клиент при этом действительно отказался от брони
def add_deposit(row):
    if row['is_canceled'] == 1 and row['is_canceled_predicted'] == 1:
        price = price_by_type[int(row.reserved_room_type)]
        date = datetime.datetime(int(row.arrival_date_year), 
                                 int(list(calendar.month_name).index(encoder.inverse_transform(row[encode_columns].values.reshape(1, -1))[0][0])),
                                 int(row.arrival_date_day_of_month))
        
        return row['revenue'] + 0.8 * (price['per_night'] * coeff_by_season[(date.month % 12) // 3] + price['service'])
    
    return row['revenue']
```

```python
# Посчитаем прибыль для каждого бронирования с внедрением депозитов и учётом бюджета на разработку системы прогнозирования
predicted_data = test_data.assign(is_canceled_predicted=best_predictions)
predicted_data['revenue_with_deposit'] = predicted_data.apply(add_deposit, axis=1)
print(f'Прибыль отеля с внедрением депозитов и учётом бюджета на разработку системы прогнозирования — {predicted_data["revenue_with_deposit"].sum() - 400000}')
```
